```python
#Recency Calculation
start_date = (1, 1, 2023)
current_date = (1, 5, 2023)

# dictionary to store the minimum recency for each person later on
min_recency = {}

# Remove non-last purchases from dataset and store it in a new dataset
seen_names = []
new_dataset = []

# for loop to remove non-last purchases from dataset
for transaction in dataset:
    if transaction["Name"] not in seen_names:
        
        #If the person is not in the seen_names list,
        # add it to the list and add the transaction to the new dataset
        seen_names.append(transaction["Name"])
        
        # add the transaction to the new dataset
        new_dataset.append(transaction)
    else:
        #If the person is in the seen_names list, check if the date of the transaction
        # is more recent than the one in the new dataset
        for new_transaction in new_dataset:
            
            # if the person is the same in both transactions
            if new_transaction["Name"] == transaction["Name"]:
                
                # in case the new transaction is more recent,
                # remove the old one and add the new one
                if new_transaction["Date"] < transaction["Date"]:
                    
                    # remove the old transaction
                    new_dataset.remove(new_transaction)
                    
                    # add the new transaction
                    new_dataset.append(transaction)

# for loop to calculate the recency of each person in dataset
for transaction in new_dataset:
    
    # get the date of each person last purchase
    last_purchase_date = transaction["Date"]
    
    # creating a future key named "Recency" to store the recency of each person
    transaction["Recency"] = 0
    
    # loop through the months between start and current date and simply do the maths manually since it's a small interval of months in the same year
    for month in range(start_date[1], current_date[1] + 1):
        if month == 1 and month == last_purchase_date[1]:
            days_in_month = 31
            transaction["Recency"] = days_in_month - last_purchase_date[0] + 1 + 28 + 31 + 30
        elif month == 2 and month == last_purchase_date[1]:
            days_in_month = 28
            transaction["Recency"] = days_in_month - last_purchase_date[0] + 1 + 31 + 30
        elif month == 3 and month == last_purchase_date[1]:
            days_in_month = 31
            transaction["Recency"] = days_in_month - last_purchase_date[0] + 1 + 30
        elif month == 4 and month == last_purchase_date[1]:
            days_in_month = 30
            transaction["Recency"] = days_in_month - last_purchase_date[0] + 1
        elif month == 5 and month == last_purchase_date[1]:
            transaction["Recency"] = 0
            
    # store the minimum recency for each person in the min_recency dictionary
    # we are doing this because otherwise we would have
    # different recency values for the same person and we just want the minimum value)
    if transaction["Name"] not in min_recency or transaction["Recency"] < min_recency[transaction["Name"]]:
        min_recency[transaction["Name"]] = transaction["Recency"]

    # print the recency of each person in dataset for visualisation
    print(f'{transaction["Name"]} is in {transaction["Recency"]} consecutive recency days!')
```

```python
# Recency Scores

# we can be sure that any recency value encountered in the for loop will be
# less than min_recency_student and greater than max_recency_student on the first iteration,
# and then update these values as necessary
# we also make sure min_recency_student is initialized to infinity
# so that any value will be less than it on the first iteration
# and it won't be 0 forever

min_recency_student = float('inf')
max_recency_student = 0

# Calculate the number of students and store the min and max recency of all students
for student, recency in min_recency.items():
    
    # Store the max and min recency of each student
    if recency > max_recency_student:
        max_recency_student = recency
    if recency < min_recency_student:
        min_recency_student = recency

# The classes are divided by the number of students and their recency
# 1 means more recency and 3 means less recency
recency_division_3 = min_recency_student + (max_recency_student - min_recency_student) / 3
recency_division_1 = max_recency_student - (max_recency_student - min_recency_student) / 3

# Create a dictionary to store the students in each class
recency_class_3 = {}
recency_class_2 = {}
recency_class_1 = {}

# Divide the students in classes based on their recency
for student, recency in min_recency.items():
    
    # If the recency of the student is between the division_1 and division_3, add them to class 2
    if recency >= recency_division_3 and recency < recency_division_1:
        recency_class_2[student] = recency
        
    # If the recency of the student is bigger than division_1, add them to class 1
    elif recency >= recency_division_1:
        recency_class_1[student] = recency
        
    # If the recency of the student is smaller than division_3, add them to class 3
    else:
        recency_class_3[student] = recency
```

```python
# Frequency Scores

# we can be sure that any frequency value encountered in the for loop will be
# less than min_frequency_student and greater than max_frequency_student on the first iteration,
# we also make sure that min_frequency_student won't be 0 setting it to infinity

max_frequency_student = 0
min_frequency_student = float('inf')

# for loop to store the min and max frequency of all students
for student, frequency in customer_freq.items():
    
    # Store the max and min frequency of each student
    if frequency > max_frequency_student:
        max_frequency_student = frequency
    if frequency < min_frequency_student:
        min_frequency_student = frequency
        
# The classes are divided by the number of students and their frequency
# 1 means more frequency and 3 means less frequency
frequency_division_3 = min_frequency_student + (max_frequency_student - min_frequency_student) / 3
frequency_division_1 = max_frequency_student - (max_frequency_student - min_frequency_student) / 3

# Create a dictionary to store the students in each class
frequency_class_3 = {}
frequency_class_2 = {}
frequency_class_1 = {}

# Divide the students in classes based on their frequency
for student, frequency in customer_freq.items():
    
    # If the frequency of the student is between the division_1 and division_3, add them to class 2
    if frequency >= frequency_division_3 and frequency < frequency_division_1:
        frequency_class_2[student] = frequency
        
    # If the frequency of the student is smaller than division_1, add them to class 1
    elif frequency <= frequency_division_1:
        frequency_class_1[student] = frequency
        
    # If the frequency of the student is bigger than division_3, add them to class 3
    else:
        frequency_class_3[student] = frequency
```

```python
# Monetary Scores

# doing the same thing as before but now for monetary 
# setting min_monetary_student to infinity so that we can be sure
# that any monetary value encountered in the for loop will be
# less than min_monetary_student and greater than max_monetary_student on the first iteration,
# so min_monetary_student won't be 0

max_monetary_student = 0
min_monetary_student = float('inf')

# for loop to store the min and max spents of all students
for student, monetary in customer_spent.items():
    
    # Store the max and min spents of each student
    if monetary > max_monetary_student:
        max_monetary_student = monetary
    if monetary < min_monetary_student:
        min_monetary_student = monetary
        
# The classes are divided by the number of students and their spents
# 1 means more spents and 3 means less spents
monetary_division_3 = min_monetary_student + (max_monetary_student - min_monetary_student) / 3
monetary_division_1 = max_monetary_student - (max_monetary_student - min_monetary_student) / 3

# Create a dictionary to store the students in each class
monetary_class_3 = {}
monetary_class_2 = {}
monetary_class_1 = {}

# Divide the students in classes based on their spents
for student, monetary in customer_spent.items():
    
    # If the spents of the student are between the division_1 and division_3, add them to class 2
    if monetary >= monetary_division_3 and monetary < monetary_division_1:
        monetary_class_2[student] = monetary
        
    # If the spents of the student are bigger than division_1, add them to class 1
    elif monetary <= monetary_division_1:
        monetary_class_1[student] = monetary
        
    # If the spents of the student are smaller than division_3, add them to class 3
    else:
        monetary_class_3[student] = monetary
```

```python
# Define the actual RFM score for each student
RFM_rank = {}

# For each student, check in which class they are in each category
for person, recency in min_recency.items():
    
    # Create a list to store the ranks of each student
    ranks = []
    
    # Check in which class the student is in each category
    # and append the rank to the list of ranks
    # this is to make it easier to compare the ranks later on
    if person in recency_class_1:
        ranks.append('1')
    elif person in recency_class_2:
        ranks.append('2')
    else:
        ranks.append('3')
    for person2, freq in customer_freq.items():
        if person == person2:
            if person in frequency_class_1:
                ranks.append('1')
            elif person in frequency_class_2:
                ranks.append('2')
            else:
                ranks.append('3')
            for person3, spent in customer_spent.items():
                if person == person3:
                    if person in monetary_class_1:
                        ranks.append('1')
                    elif person in monetary_class_2:
                        ranks.append('2')
                    else:
                        ranks.append('3')
                    RFM_rank[person] = ranks

mylist = list(RFM_rank.items())
sorted_list = []
while len(mylist) != 0:
    
    # Set the max rank to -1
    max_rank = ["-1", "-1", "-1"]
    max_rank_student = ""
    
    # For each student, check their rank and compare it to the max rank
    for student, rank in mylist:
        
        # If the rank of the student is bigger than the max rank, set the max rank to the student's rank
        if rank > max_rank:
            max_rank = rank
            max_rank_student = student
            
        # Append the student to the sorted list
    sorted_list.append((max_rank_student, max_rank))
    mylist.remove((max_rank_student, max_rank))

# Print the sorted list
for student, rank in sorted_list:
    print(student, rank)
```

### Alphabetical Sorting

```python
# base code for merge sort
def sort_merge(list_of_names):
    
    # if the list of names has only one name in it, return it
    if len(list_of_names) == 1:
        return list_of_names
    
    # if the list of names has more than one name in it, split it in half
    idx_half = len(list_of_names) // 2
    
    # sort the left half of the list
    # sort the right half of the list
    list_of_names_left = sort_merge(list_of_names[:idx_half])
    list_of_names_right = sort_merge(list_of_names[idx_half:])
    
    # merge the two halves of the list
    return merge(list_of_names_left, list_of_names_right)

# base code for merge function
def merge(list_of_names_left, list_of_names_right):
    
    # create an empty list to store the sorted names
    list_of_names_sorted = []
    
    # create two variables to keep track of the index of the left and right lists
    # set them to 0
    l, r = 0, 0
    
    # while the index of the left list is smaller than the length of the left list
    while l < len(list_of_names_left) and r < len(list_of_names_right):
        
        # if the name in the left list is smaller than the name in the right list
        if list_of_names_left[l] < list_of_names_right[r]:
            
            #append the name in the left list to the sorted list
            list_of_names_sorted.append(list_of_names_left[l])
            l += 1
            
        # if the name in the right list is smaller than the name in the left list
        else:
            
            # append the name in the right list to the sorted list
            list_of_names_sorted.append(list_of_names_right[r])
            r += 1
            
    # if the index of the left list is smaller than the length of the left list
    if l < len(list_of_names_left):
        
        # append the rest of the names in the left list to the sorted list
        list_of_names_sorted.extend(list_of_names_left[l:])
        
    # if the index of the right list is smaller than the length of the right list
    if r < len(list_of_names_right):  
        
        # append the rest of the names in the right list to the sorted list
        list_of_names_sorted.extend(list_of_names_right[r:])
        
    # return the sorted list
    return list_of_names_sorted

# print the sorted list
alphabetic_names=sort_merge((list(min_recency.keys())))
print(alphabetic_names)
```

```python
# base code for binary search
def binary_search(list_names, name):
    
    # the left is the index of the first letter of the name
    # the right is the index of the last letter of the name
    left = 0
    right = len(list_names) - 1 
    
    # while the index of the first letter of the name is
    # smaller than the index of the last letter of the name
    while left <= right:
        
        # the middle is the index of the middle letter of the name
        mid = (left + right) // 2
        
        # if the middle letter of the name is the letter you're looking for
        if list_names[mid] == name:
            
            # print that the name was found and in what position
            # We incremented 1 to the index because the index starts at 0
            # We hope it is fine anyway :)
            return mid
        
        # if the middle letter of the name is smaller than the letter you're looking for
        # the left is now the index of the middle letter of the name + 1
        # (since the middle letter of the name is already checked)
        elif list_names[mid] < name:
            left = mid + 1
            
        # if the middle letter of the name is bigger than the letter you're looking for
        # the right is now the index of the middle letter of the name - 1
        # (since the middle letter of the name is already checked)    
        else:
            right = mid - 1
            
    # if the name isn't in the list, print that the name wasn't found        
    # and return None to make sure it doesn't return anything
    return None

# base code for print position
# so it isn't inside the algorithm and it's easier to read
def print_position(person_name):
    position = binary_search(alphabetic_names, person_name)
    
    if position is not None:
        print(f"You're in position {position + 1}, {person_name}. YOU'RE HERE!")
    else:
        print(f"Who is {person_name}? Are they an IMS student? Or are they FROM ANOTHER DIMENSION?")    

# printing to make sure it works
print_position('Pedro')
print_position('Filipa')
print_position('Beatris')

# to show it works here's what happens when you search for someone who isn't in the dataset
print_position('Soo')
```
